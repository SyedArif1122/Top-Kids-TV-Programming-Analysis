# **Question 1**: Top channels with highest average total viewers

```python
top_channels = df.groupby('Channel')['Total Live Viewers'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_channels.values, y=top_channels.index, palette="viridis")
plt.xlabel("Average Total Viewers (in Hundreds)")
plt.ylabel("Channel")
plt.title("Top Channels with Highest Average Total Viewers")
plt.show()
```

# Question 2: Episode or series with highest total viewers

```python
highest_viewers_episode = df[df['Total Live Viewers'] == df['Total Live Viewers'].max()]
print("Episode or series with highest total viewers:\n", highest_viewers_episode)
```

# Question 3: Correlation between release year and total viewership

```python
correlation = df['Release Year'].corr(df['Total Live Viewers'])
correlation
```

# Question 4: Distribution of total viewership across channels

```python
channel_distribution = df['Channel'].value_counts()
plt.figure(figsize=(10, 6))
sns.countplot(y="Channel", data=df, palette="viridis")
plt.xlabel("Number of Episodes")
plt.ylabel("Channel")
plt.title("Distribution of Total Viewership Across Channels")
plt.show()
```

# Question 5: Trends in total viewership over the years

```python
total_viewers_trends = df.groupby('Release Year')['Total Live Viewers'].sum()
plt.figure(figsize=(10, 6))
total_viewers_trends.plot(marker='o', color='b')
plt.xlabel("Release Year")
plt.ylabel("Total Viewers (in Hundreds)")
plt.title("Trends in Total Viewership Over the Years")
plt.grid(True)
plt.show()
```

# Question 6: Most popular series among kids based on total viewers

```python
popular_series = df.groupby('Series')['Total Live Viewers'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_series.values, y=popular_series.index, palette="viridis")
plt.xlabel("Total Viewers (in Hundreds)")
plt.ylabel("Series")
plt.title("Most Popular Series Among Kids Based on Total Viewers")
plt.show()
```

# Question 7: Seasonal pattern in total viewership

```python
seasonal_pattern = df.groupby(df['Release Year'] // 10 * 10)['Total Live Viewers'].mean()
plt.figure(figsize=(10, 6))
seasonal_pattern.plot(marker='o', color='g')
plt.xlabel("Release Decade")
plt.ylabel("Average Total Viewers (in Hundreds)")
plt.title("Seasonal Pattern in Total Viewership")
plt.grid(True)
plt.show()
```
